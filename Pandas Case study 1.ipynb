```python
value_segment=[]
for i in range(len(customer)):
    if customer.Customer_Value[i]>25000:
        value_segment.append('High')
    elif customer.Customer_Value[i]<=10000:
        value_segment.append('Low')
    else:
        value_segment.append('Medium')
customer['Customer_Value_Segment']=value_segment
customer.columns
```

```python
customer['Avg_Revenue_per_trip']=customer['Customer_Value']/customer['Buy_Times']
customer['Balance_points']=customer['Points_Earned']-customer['Points_Redeemed']
customer.columns
```

```python
# difference in last purchase and today

from datetime import date

today = date.today()
print(f"Today's date:::{today}")
cust_hist={}
for i in range(len(customer)):
    dt=str(customer.Recent_Date[i])
    x=date(int(dt[:4]), int(dt[4:6]), int(dt[6:]))-today
    c=customer.Customer_ID[i]
    cust_hist[c]=x.days*-1
cust_hist
```

```python
region=customer.Last_region.unique()
state=customer.Last_state.unique()
city=customer.Last_city.unique()
```

```python
region
```

```python
total=customer.Customer_Value.sum()
print(f"Total worth:::{total}")
region_perc={}
state_perc={}
city_perc={}
for r in region:
    temp=customer[customer.Last_region==r].Customer_Value.sum()
    region_perc[r]=temp*100/total

for s in state:
    temp=customer[customer.Last_state==s].Customer_Value.sum()
    state_perc[s]=temp*100/total
    
for c in city:
    temp=customer[customer.Last_city==c].Customer_Value.sum()
    city_perc[c]=temp*100/total
```

```python
state_count={}
city_count={}
state_avg_purchase={}
city_avg_purchase={}

total_count=customer.Customer_ID.count()
total_purchase=customer.Buy_Times.sum()
```

```python
for s in state:
    temp=customer[customer.Last_state==s].Customer_ID.count()
    state_count[s]=temp
    temp=customer[customer.Last_state==s].Buy_Times.sum()
    state_avg_purchase[s]=temp*100/total_purchase
    
for c in city:
    temp=customer[customer.Last_city==c].Customer_ID.count()
    city_count[c]=temp
    temp=customer[customer.Last_city==c].Buy_Times.sum()
    city_avg_purchase[c]=temp*100/total_purchase
```
